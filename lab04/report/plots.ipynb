```python
from kmeans import kmeans, kmeans_plot
from hclustering import hclust, hclust_plot

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# 4clusters.csv

```python
fourclust = pd.read_csv("data/4clusters.csv", skiprows=1, header=None)
```

```python
kmeans_plot(kmeans(fourclust, 4))
```

```python
hclust_plot(hclust(fourclust, 4, by="nclust"))
```

# Planets

```python
planets = pd.read_csv('data/planets.csv', header=None, skiprows=1)
planets.drop(0, axis=1, inplace=True)
planets.columns = range(len(planets.columns))
```

```python
kmeans_plot(kmeans(planets, 3))
ax = plt.gca()
ax.dist = 12
ax.set_xlabel("Node (Degrees)")
ax.set_ylabel("Inclination (Degrees)")
ax.set_zlabel("Axis (Astronomic Units)")
plt.show()
```

```python
hclust_plot(hclust(planets, 3, by="nclust"))
ax = plt.gca()
ax.dist = 12
ax.set_xlabel("Node (Degrees)")
ax.set_ylabel("Inclination (Degrees)")
ax.set_zlabel("Axis (Astronomic Units)")
plt.show()
```

# Iris

```python
iris = pd.read_csv('data/iris.csv', skiprows=1, header=None)
iris.drop(4, axis=1, inplace=True)
```

```python
kmeans_plot(kmeans(iris, 3), cols=[0,1])
ax = plt.gca()
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
plt.show()
```

```python
kmeans_plot(kmeans(iris, 3), cols=[2,3])
ax = plt.gca()
ax.set_xlabel("Petal Length")
ax.set_ylabel("Petal Width")
plt.show()
```

```python
hclust_plot(hclust(iris, 3, by="nclust"), cols=[0,1])
ax = plt.gca()
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
plt.show()
```

```python
hclust_plot(hclust(iris, 3, by="nclust"), cols=[2,3])
ax = plt.gca()
ax.set_xlabel("Petal Length")
ax.set_ylabel("Petal Width")
plt.show()
```
