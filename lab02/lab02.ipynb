### 1k Rows

```python
pd.read_csv("https://users.csc.calpoly.edu/~dekhtyar/466-Fall2023/data/BAKERY/1000/1000-out2.csv", header=None, index_col=0).sum(axis=0).sort_values().divide(1000).plot.bar()
```
